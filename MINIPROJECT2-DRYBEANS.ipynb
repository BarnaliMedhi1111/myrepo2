 # 1.Understanding the Problem Statement
 
##  Dataset Information :

 Seven different types of dry beans were used in this project, taking into account the  features such as form, shape, type and structure by the market situation. We have to use  best ML alogorithm to classify the most well-known 7 types  of beans in Turkey, Barbunya, Bombay, Cali, Dermason, Horoz, Seker and Sira depending only on dimension and shape features of bean varieties with no external discriminatory features.
##   Features Information :
 1. Area (A) : The area of a bean zone and the number of pixels within its boundaries.
 2. Perimeter (P) : Bean circumference is defined as the length of it's border.
 3. Major axis length (L) : The distance between the ends of the longest line that can be drawn from a bean.
 4. Minopr axis length (I) : The longest line that can drawn from the bean while standing perpendicular to the main axis.
 5. Aspect ratio (K) : Defines the relationship between L and I.
 6. Eccentricity (Ec) : Eccentricity of the elipse having the same moments as the region.
 7. Cnovex area (C) : Number of pixel in the smallest convex polygon that can contain the area of a bean seed.
 8. Equivalent diameter (Ed) : The diameter of a circle having the same area as a bean seed area.
 9. Extent (Ex) : The ratio of the pixels in the bounding box to the bean area.
 10. Solidity (S) : Also known as convexity. The ratio of the pixels in the convex shell to those found in beans.
 11. Roundness (R) : Calculated with the the following formula : (4piA)/(P^2)
 12. Compactness (CO) : Measures the roundness of an object : Ed/L
 13. ShadeFactor (SF1)
 14. ShapeFactor (SF2)
 15. ShapeFactor (SF3)
 16. ShapeFactor (SF4)
 17. Class (Seker, Barbunya, Cali, Dermosan, Horoz and Sira)

# 5.Model Training

I've tried SVM & with SVM with HYPER_PARAMETER_TUNING,DecisionTree but best results are given by Rndom_Forest

so i'll be submitting only Random Forest

```python
# importnat package for building model
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

## Splitting Training and testing data set

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=23)
#keeping training size 70%
#testing size 30%
```

```python
print("X_train_shape",X_train.shape)
print("X_test_shape",X_test.shape)
print("y_train_shape",y_train.shape)
print("y_test_shape",y_test.shape)
```

### Model

```python
#initializing model
rfc=RandomForestClassifier(random_state=23)

#fitting data 
rfc.fit(X_train,y_train)
```

### Prediction

```python
y_pred_rfc=rfc.predict(X_test)
y_pred_rfc
```

### Checking testing score

```python
print(rfc.score(X_test,y_test))
```

### Training score

```python
print(rfc.score(X_train,y_train))
```

### Prediction Score

```python
print(rfc.score(X_test,y_pred_rfc))
```

### Comparing Actual and Predicted score

```python
#checking score with metrics
from sklearn.metrics import accuracy_score
score_rfc=accuracy_score(y_test,y_pred_rfc)
print("comparision score between actual and predicted labels:",score_rfc)
```
